```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns 
%matplotlib inline 

import warnings 
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('data.csv',error_bad_lines = False)
data.head()
```

```python
data['strength'].unique() 
#0 is poor, 1 medium, 2 is strongest
```

```python
data.isna().sum() #password has 1 null value and strength has no null value , hence only 1 null value could be dropped
```

```python
data[data['password'].isna()] #checking which row has the null value of password
```

```python
data.dropna(inplace = True) #dropping all the null values, however we only have one
```

```python
data.isnull().sum()
```

```python
sns.countplot(data['strength'])
#the password with 1 strength(medium) has count of 50000 whereas the others are somewhere close to 10000
#the dataset is imbalanced
#password is the independent feature , strength is the dependent feature
```

```python
#creating the entire data into an array as password_tuple
password_tuple = np.array(data)
```

```python
password_tuple
```

```python
#shuffling the data so that it provides robustness to the model
import random
random.shuffle(password_tuple)
```

```python
x = [labels[0] for labels in password_tuple] #password has index 0 in the dataset
y = [labels[1] for labels in password_tuple] #strength has index 1 in the dataset
```

```python
x
```

```python
y
```

```python
def word_divide_char(inputs):
  character = [] #this is a list in which the letters get stored
  for i in inputs:
    character.append(i)
  return character
```

```python
word_divide_char('kzde5577')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
vectorizer = TfidfVectorizer(tokenizer=word_divide_char)
```

```python
X = vectorizer.fit_transform(x)
```

```python
X.shape
```

```python
vectorizer.get_feature_names() #gives the features in our data 
```

```python
#now finding the importance of each feature in our data
```

```python
first_document_vector = X[0]
first_document_vector #this is the object of the vector or the sparse matrix which is used to speed up the process and save memory
```

```python
first_document_vector.T.todense()  
#the above code finally gives us the data having its own importance
# .T is used for transpose
```

```python
#We have to prepare our data for building the model which would be having the first column of the feature and the second column besides it as the importance of that same feature 
```

```python
df = pd.DataFrame(data = first_document_vector.T.todense(),index = vectorizer.get_feature_names(), columns = ['TF-IDF'] )
```

```python
df
```

```python
df.sort_values(by = 'TF-IDF',ascending = False)
```

```python
#Now our data is ready 
#Next step is to build the logistic regression model from this data
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2) 
```

```python
X.shape # this is the original data, refer above X is our whole  data
```

```python
X_train.shape #this is the data we're training as 0.2 is test size means 20% is tested hence, 80% data is trained which can be observed below
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression(random_state= 0, multi_class= 'multinomial') 
#multinomial - when we have a classification problem but we have more than 2 categories(other than 0 and 1), here we have 3 categories(0,1,2) so multinomial
```

```python
clf.fit(X_train,y_train)
```

```python
dt = np.array(['Anuj@167']) #we're storing a random password in an array for prediction
pred = vectorizer.transform(dt) #our own created vectorizer transforms the password into a format understable by the machine as machine can't understand string data
clf.predict(pred) #clf is our logistic model so the prediction takes place, obviously!!
```

```python
#the result above shows that the password is of average strength i.e 1
```

```python
y_pred = clf.predict(X_test) #testing our test data which is X_test(20% of data is being tested)
y_pred
```

```python
#We can also use advanced classifiers such as any boost classifiers(ada boost, XG boost, etc)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
```

```python
print(confusion_matrix(y_test,y_pred))
print(classification_report(y_test,y_pred))
print(accuracy_score(y_test,y_pred))
```
